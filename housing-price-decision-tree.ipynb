```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
```

```python
#getting data
os.listdir("/kaggle/input/competitions/home-data-for-ml-course")
```

```python
##getting data
house_data =pd.read_csv("/kaggle/input/competitions/home-data-for-ml-course/train.csv")
house_data_test =pd.read_csv("/kaggle/input/competitions/home-data-for-ml-course/test.csv")
```

```python
house_data.head()
```

```python
#separate target and features
y=house_data.SalePrice 
x =house_data.drop("SalePrice", axis=1)
```

```python
#split the data
train_x,val_x,train_y,val_y=train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=1)
```

```python
#preprocessing the data
numeric__train_data =train_x.select_dtypes(include=["number"])
categorical_train_data =train_x.select_dtypes(exclude=["number"])
numeric__val_data =val_x.select_dtypes(include=["number"])
categorical_val_data =val_x.select_dtypes(exclude=["number"])
```

```python
num_imputer=SimpleImputer()
cat_imputer =SimpleImputer(strategy="most_frequent")
imputed_num_train_data =pd.DataFrame(num_imputer.fit_transform(numeric__train_data ))
imputed_num_val_data=pd.DataFrame(num_imputer.transform(numeric__val_data))
imputed_cat_train_data=pd.DataFrame(cat_imputer.fit_transform(categorical_train_data))
imputed_cat_val_data =pd.DataFrame(cat_imputer.transform(categorical_val_data))
```

```python
#handling categorical variables





# Apply ordinal encoder to each column with categorical data
ordinal_encoder = OrdinalEncoder(  handle_unknown="use_encoded_value",unknown_value=-1)
label_categorical_train_data=pd.DataFrame(ordinal_encoder.fit_transform(imputed_cat_train_data))
label_categorical_val_data=pd.DataFrame(ordinal_encoder.transform(imputed_cat_val_data))
```

```python
preprocessed_train_data =pd.concat([imputed_num_train_data ,label_categorical_train_data],axis=1)
preprocrssed_val_data=pd.concat([imputed_num_val_data,label_categorical_val_data],axis=1)
```

```python
#creating a model 
house_model=DecisionTreeRegressor(random_state=1)
house_model.fit(preprocessed_train_data,train_y)
```

```python
house_prediction=house_model.predict(preprocrssed_val_data)
```

```python
mean_absolute_error(val_y,house_prediction)
```

**using Pipeline**

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
```

```python
numeric_columns =numeric__train_data.columns.tolist()
categorical_columns =categorical_train_data.columns.tolist()
```

```python
#preprocess numeric data
numeric_transformer = SimpleImputer(strategy ="constant")

#preprocessing categorical variable
categorical_transformer =Pipeline(steps=[('imputer',SimpleImputer(strategy="most_frequent")),
                                        ('onehot',OneHotEncoder(handle_unknown="ignore"))])
# Bundle preprocessing for numerical and categorical data
preprocessor =ColumnTransformer(transformers=[('num',numeric_transformer,numeric_columns),('cat',categorical_transformer,categorical_columns)  ])
```

```python
model =DecisionTreeRegressor(random_state=1)
```

```python
my_pipeline =Pipeline(steps=[("preprocessor",preprocessor),('model',model)])

my_pipeline.fit(train_x,train_y)

preds=my_pipeline.predict(val_x)

mean_absolute_error(val_y,preds)
```
